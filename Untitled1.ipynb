```python
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python

link = 'https://raw.githubusercontent.com/giguachad/portfolio/main/df4jupyter.csv' # The shareable link
df = pd.read_csv(link)
```

```python

df = df.astype({"Chg.":"str"})
df["Chg."] = df["Chg."].str.replace("+","")
for i in (df.columns):
  df = df.astype({i:"str"})
  df[i] = df[i].str.replace("%","")
  df[i] = df[i].str.replace("+","")
```

```python
df = df.drop(["Chg.", "Chg.%", "Vol."], axis = 1)
```

```python
df.head()
```

```python
#for p in df.columns:
 # if p != "Name" and p != "Last" and df.iloc[0,2] == 0.84:
  #  df[p] = df["Last"] - df["Last"]*(df[p]/100)
df = df.drop(196)
  
for j in df.columns:
  if j != "Name":
    df[j] = df[j].apply(float)

df.head(-5)
```

```python

X = df.drop("Name", axis = 1)
X_D_W = X.drop(["Month_1", "YTD", "Year_1", "Years_3"], axis = 1)
X_M_YTD = X.drop(["Daily", "Week_1", "Year_1", "Years_3"], axis = 1)
X_Y_Y = X.drop(["Daily", "Week_1", "Month_1", "YTD"], axis = 1)
```

```python
scaler = StandardScaler()

labels = []
#scores = []

scaled_features = scaler.fit_transform(X_D_W)
kmeans = KMeans(n_clusters=3, random_state = 42)
kmeans.fit(scaled_features)
labels = kmeans.labels_
score = silhouette_score(scaled_features, labels) 

df["DWLabel"] = labels
sns.scatterplot(data = df, x = "Last", y = "Month_1", hue = "DWLabel").set(title = "DWLabel vs Last Price")

score
```

```python
for i in range(497):
  if i != 196:  
    if df["DWLabel"][i] == 2:
      df["DWLabel"][i] = "LpHi"
    elif df["DWLabel"][i] == 0:
      df["DWLabel"][i] = "LpLi"
    elif df["DWLabel"][i] == 1:
      df["DWLabel"][i] = "HpHi"
df["DWLabel"] = df["DWLabel"].astype("category")
df.head()
```

```python
scaler1 = StandardScaler()

labels = []
#scores = []

scaled_features = scaler1.fit_transform(X_M_YTD)
kmeans1 = KMeans(n_clusters=3, random_state = 42)
kmeans1.fit(scaled_features)
labels = kmeans1.labels_
score = silhouette_score(scaled_features, labels) 

df["MYTDLabel"] = labels
sns.scatterplot(data = df, x = "Last", y = "YTD", hue = "MYTDLabel").set(title = "YTD vs Last Price")
score
```

```python
for i in range(497):
  if i != 196:
    if df["MYTDLabel"][i] == 1:
      df["MYTDLabel"][i] = "HpHi"
    elif df["MYTDLabel"][i] == 0:
      df["MYTDLabel"][i] = "LpHi"
    elif df["MYTDLabel"][i] == 2:
      df["MYTDLabel"][i] = "LpLi"
df["MYTDLabel"] = df["MYTDLabel"].astype("category")
df.head()
```

```python
scaler2 = StandardScaler()

labels = []
#scores = []

scaled_features = scaler2.fit_transform(X_Y_Y)
kmeans2 = KMeans(n_clusters=3, random_state = 42)
kmeans2.fit(scaled_features)
labels = kmeans2.labels_
score = silhouette_score(scaled_features, labels) 

df["MYYLabel"] = labels
sns.scatterplot(data = df, x = "Last", y = "Year_1", hue = "MYYLabel").set(title = "MYYLabel vs Last Price")
score
```
